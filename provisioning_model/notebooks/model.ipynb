```python
from provisioning_model.utils.image import concatenate_images_with_centered_below, save_erdantic_diagram

diagrams = [
    extraction_diagram,
    transformation_diagram,
    dissipation_diagram,
]
diagrams_simple = [
    extraction_diagram_1,
    transformation_diagram_1,
    dissipation_diagram_1,
]
output_path = 'provisioning.png'
output_path_simple = 'provisioning_simple.png'

img_paths = []
img_paths_simple = []
counter = 0
counter_simple = 0

for diagram in diagrams:
    img_path = save_erdantic_diagram(diagram, f'provisioning_{counter}')
    img_paths.append(img_path)
    counter += 1

for diagram_simple in diagrams_simple:
    img_path = save_erdantic_diagram(diagram_simple, f'provisioning_simple_{counter_simple}')
    img_paths_simple.append(img_path)
    counter_simple += 1

centered_below_path = save_erdantic_diagram(decision_making_diagram, 'provisioning_decision_making')
centered_below_path_simple = save_erdantic_diagram(decision_making_diagram_1, 'provisioning_decision_making_simple')

concatenate_images_with_centered_below(img_paths, centered_below_path, output_path)
concatenate_images_with_centered_below(img_paths_simple, centered_below_path_simple, output_path_simple)
```
