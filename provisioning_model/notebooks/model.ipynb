```python
from dataclasses import dataclass
import erdantic as erd
```

```python
from provisioning_model.model.extraction.input import Employment, EnergyUse, CapitalUse, ResourceUse
from provisioning_model.model.extraction.exchange_incoming import Imports
```

```python
from provisioning_model.model.transformation.efficiency import Efficiency
from provisioning_model.model.transformation.structure import Structure
from provisioning_model.model.transformation.infrastructure import Infrastructure
```

```python
from provisioning_model.model.dissipation.output import NeedSatisfaction, Accumulation, HumanDevelopment, \
    WasteAndPollution, \
    ImpactOnEcosystems
from provisioning_model.model.dissipation.exchange_outgoing import Exports
```

```python
@dataclass
class Extraction:
    energy_use: EnergyUse
    employment: Employment
    resource_use: ResourceUse
    capital_use: CapitalUse
    imports: Imports


@dataclass
class Transformation:
    efficiency: Efficiency
    structure: Structure
    infrastructure: Infrastructure


@dataclass
class Dissipation:
    need_satisfaction: NeedSatisfaction
    accumulation: Accumulation
    human_development: HumanDevelopment
    waste_and_pollution: WasteAndPollution
    impact_on_ecosystems: ImpactOnEcosystems
    exports: Exports
```

```python
extraction_diagram = erd.create(Extraction)
extraction_diagram
```

```python
transformation_diagram = erd.create(Transformation)
transformation_diagram
```

```python
dissipation_diagram = erd.create(Dissipation)
dissipation_diagram
```

```python
from provisioning_model.model.software_governance.state import Government, CorporatePower
from provisioning_model.model.software_governance.foreign_relations import ForeignRelations
from provisioning_model.model.software_governance.culture import Culture
from provisioning_model.model.software_governance.communication import MediaStructure
from provisioning_model.model.software_governance.inequality import EconomicInequality, GenderInequality


@dataclass
class Inequality:
    economic_inequality: EconomicInequality
    gender_inequality: GenderInequality


@dataclass
class SoftwareGovernance:
    government: Government
    corporate_power: CorporatePower
    foreign_relations: ForeignRelations
    culture: Culture
    media_structure: MediaStructure
    inequality: Inequality


decision_making_diagram = erd.create(SoftwareGovernance)
decision_making_diagram
```

```python
@dataclass
class Provisioning:
    extraction: Extraction
    transformation: Transformation
    dissipation: Dissipation
    software_governance: SoftwareGovernance
```

```python
from provisioning_model.utils.image import concatenate_images_horizontally, save_erdantic_diagram

diagrams = [
    decision_making_diagram,
    extraction_diagram,
    transformation_diagram,
    dissipation_diagram,
]
output_path = 'provisioning.png'

img_paths = []
counter = 0
for diagram in diagrams:
    img_path = save_erdantic_diagram(diagram, f'provisioning_{counter}')
    img_paths.append(img_path)
    counter += 1

concatenate_images_horizontally(img_paths, output_path)
```
