```python
@dataclass
class Extraction:
    worker_fund: WorkerFund
    capital_fund: CapitalFund
    energy_stock: EnergyStock
    resource_stock: ResourceStock
    imports: Imports


@dataclass
class IndustryStructure:
    industry: Industry
    agriculture: Agriculture
    services: Services


@dataclass
class FoundationalStructure:
    foundational_economy: FoundationalEconomy
    overlooked_economy: OverlookedEconomy
    core_economy: CoreEconomy
    competitive_economy: CompetitiveEconomy


@dataclass
class ProvisioningRealms:
    state_realm: StateRealm
    market_realm: MarketRealm
    household_realm: HouseholdRealm
    commons_realm: CommonsRealm


@dataclass
class Structure:
    industry_structure: IndustryStructure
    foundational_structure: FoundationalStructure
    provisioning_realms: ProvisioningRealms


@dataclass
class Transformation:
    efficiency: Efficiency
    structure: Structure
    infrastructure: Infrastructure


@dataclass
class Dissipation:
    need_satisfaction: NeedSatisfaction
    accumulation: Accumulation
    human_development: HumanDevelopment
    waste_and_pollution: WasteAndPollution
    exports: Exports
```

```python
extraction_diagram = erd.create(Extraction)
extraction_diagram
```

```python
transformation_diagram = erd.create(Transformation)
transformation_diagram
```

```python
dissipation_diagram = erd.create(Dissipation)
dissipation_diagram
```

```python
# from provisioning_model.model.software_governance.foreign_relations import ForeignRelations
# from provisioning_model.model.software_governance.culture import MediaStructure, LegalSystem
# from provisioning_model.model.software_governance.inequality import EconomicInequality, GenderInequality

class ForeignRelations: pass


class MediaStructure: pass


class LegalSystem: pass


class EconomicInequality: pass


class GenderInequality: pass


@dataclass
class Inequality:
    economic_inequality: EconomicInequality
    gender_inequality: GenderInequality


@dataclass
class Culture:
    media_structure: MediaStructure
    legal_system: LegalSystem


@dataclass
class SoftwareGovernance:
    foreign_relations: ForeignRelations
    culture: Culture
    inequality: Inequality


decision_making_diagram = erd.create(SoftwareGovernance)
decision_making_diagram
```

```python
@dataclass
class Provisioning:
    extraction: Extraction
    transformation: Transformation
    dissipation: Dissipation
    software_governance: SoftwareGovernance
```

```python
from provisioning_model.utils.image import concatenate_images_with_centered_below, save_erdantic_diagram

diagrams = [
    extraction_diagram,
    transformation_diagram,
    dissipation_diagram,
]
output_path = 'provisioning.png'

img_paths = []
counter = 0
for diagram in diagrams:
    img_path = save_erdantic_diagram(diagram, f'provisioning_{counter}')
    img_paths.append(img_path)
    counter += 1
centered_below_path = save_erdantic_diagram(decision_making_diagram, 'provisioning_decision_making')

concatenate_images_with_centered_below(img_paths, centered_below_path, output_path)
```
